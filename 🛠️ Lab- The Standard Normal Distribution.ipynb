## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
# Calculate the mean and standard deviation
mean_yield = data['0'].mean()
std_yield = data['0'].std()

# Calculate the interval bounds for 99% of the observations
lower_bound = mean_yield - 2.576 * std_yield
upper_bound = mean_yield + 2.576 * std_yield

print(f"99% of the observations fall between {lower_bound:.2f} and {upper_bound:.2f} pounds.")
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
# Calculate the mean and standard deviation
mean_yield = data['0'].mean()
std_yield = data['0'].std()

# Given value
X = 35

# Calculate the z-score
z_score = (X - mean_yield) / std_yield

print(f"The z-score for a tree yielding 35 pounds of apples is {z_score:.2f}.")
```

# Interpret the result
The z-score for a tree yielding 35 pounds of apples is -1.23.


## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

```python
# Interpret the z score
# Calculate the mean and standard deviation
mean_yield = data['0'].mean()
std_yield = data['0'].std()

# Given z-score
z_score = 1.85

# Calculate the yield
yield_value = z_score * std_yield + mean_yield

print(f"The yield of the tree with a z-score of 1.85 is {yield_value:.2f} pounds.")
```

# What is the yield ?
The yield of the tree with a z-score of 1.85 is 53.51 pounds.
